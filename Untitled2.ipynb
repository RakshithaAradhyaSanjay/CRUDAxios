```python
import matplotlib.pyplot as plt
import numpy as np

# Data
t = np.arange(0., 5., 0.2)

# Create the plot
plt.plot(t, t, 'ro', label='t vs t')       # Red circles
plt.plot(t, t**2, 'bs', label='t vs t^2')  # Blue squares
plt.plot(t, t**4, 'g^', label='t vs t^4')  # Green triangles

# Set axis limits
plt.xlim([0, 6])
plt.ylim([0, 150])

# Add labels and title
plt.xlabel('t')
plt.ylabel('Values')
plt.title('Multiple Lines on the Same Graph')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
bar_width = 0.3
bars1 = [10, 9, 2]
bars2 = [10.8, 9.5, 4.5]
yer1 = [0.5, 0.4, 0.5]
yer2 = [1, 0.7, 1]
r1 = np.arange(len(bars1))
r2 = [x + bar_width for x in r1]

# Create the first bar plot
plt.bar(r1, bars1, width=bar_width, color='blue', edgecolor='black', yerr=yer1, capsize=7, label='Female')

# Create the second bar plot
plt.bar(r2, bars2, width=bar_width, color='cyan', edgecolor='black', yerr=yer2, capsize=7, label='Male')

# General layout
plt.xticks([r + bar_width / 2 for r in range(len(bars1))], ['Kids', 'Adults', 'Older people'])
plt.title('Advanced bar plot')
plt.ylabel('height')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Data
np.random.seed(123)
mu = 120  # mean of distribution
sigma = 12  # standard deviation of distribution
x = mu + sigma * np.random.randn(10000)
num_bins = 20

# Create histogram
n, bins, patches = plt.hist(x, num_bins, density=1, facecolor='red', alpha=0.6)

# Add a 'best fit' line
y = norm.pdf(bins, mu, sigma)
plt.plot(bins, y, 'b--')

# Customizations
plt.title("Histogram with fitted line")
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
np.random.seed(123)
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
x = np.concatenate((spread, center, flier_high, flier_low))

green_diamond = dict(markerfacecolor='g', marker='D')

# Create box plot
plt.boxplot(x, flierprops=green_diamond)

# Customizations
plt.title("My box plot")
plt.xlabel("My data")

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse

# Data
df = pd.read_csv('https://github.com/selva86/datasets/raw/master/AirPassengers.csv')
dates = pd.DatetimeIndex([parse(d).strftime('%Y-%m-01') for d in df['date']])
df.set_index(dates, inplace=True)
ts = df['value']

# Create time series plot
ts.plot()

# Customizations
plt.title('My time series')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')

# Show plot
plt.show()
```
